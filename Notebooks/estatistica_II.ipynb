## ‚úÖ 2.1 Distribui√ß√£o Binomial
***

Um evento **binomial** √© caracterizado pela possibilidade de ocorr√™ncia de apenas duas categorias. Estas categorias somadas representam todo o espa√ßo amostral, sendo tamb√©m mutuamente excludentes, ou seja, a ocorr√™ncia de uma implica na n√£o ocorr√™ncia da outra.

Em an√°lises estat√≠sticas o uso mais comum da distribui√ß√£o binomial √© na solu√ß√£o de problemas que envolvem situa√ß√µes de **sucesso** e **fracasso**.

# $$P(k)=\binom{n}{k} p^k q^{n-k}$$

Onde:

$p$ = probabilidade de sucesso

$q = (1 - p)$ = probabilidade de fracasso

$n$ = n√∫mero de eventos estudados

$k$ = n√∫mero de eventos desejados que tenham sucesso

### Experimento Binomial

1. Realiza√ß√£o de $n$ ensaios id√™nticos.

2. Os ensaios s√£o independentes.

3. Somente dois resultados s√£o poss√≠veis, exemplo: Verdadeiro ou falso; Cara ou coroa; Sucesso ou fracasso.

4. A probabilidade de sucesso √© representada por $p$ e a de fracasso por $1-p=q$. Estas probabilidades n√£o se modificam de ensaio para ensaio.

### M√©dia da distribui√ß√£o binomial

O valor esperado ou a m√©dia da distribui√ß√£o binomial √© igual ao n√∫mero de experimentos realizados multiplicado pela chance de ocorr√™ncia do evento.

# $$\mu = n \times p$$

### Desvio padr√£o da distribui√ß√£o binomial

O desvio padr√£o √© o produto entre o n√∫mero de experimentos, a probabilidade de sucesso e a probabilidade de fracasso.

# $$\sigma = \sqrt{n \times p \times q}$$

### Importando bibliotecas
https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.comb.html

```python
from scipy.special import comb
```

### Combina√ß√µes

N√∫mero de combina√ß√µes de $n$ objetos, tomados $k$ a cada vez, √©:

# $$C_{k}^{n} = \binom{n}{k} = \frac{n!}{k!(n - k)!}$$

Onde

## $$n! = n\times(n-1)\times(n-2)\times...\times(2)\times(1)$$
## $$k! = k\times(k-1)\times(k-2)\times...\times(2)\times(1)$$

Por defini√ß√£o

## $$0! = 1$$

## üìå Exemplo: Mega Sena

Em um volante de loteria da Mega Sena temos um total de **60 n√∫meros** para escolher onde a aposta m√≠nima √© de **seis n√∫meros**. ## üìå Exemplo: Concurso para cientista de dados

Em um concurso para preencher uma vaga de cientista de dados temos um total de **10 quest√µes** de m√∫ltipla escolha com **3 alternativas poss√≠veis** em cada quest√£o. **Cada quest√£o tem o mesmo valor.** Suponha que um candidato resolva se aventurar sem ter estudado absolutamente nada. Ele resolve fazer a prova de olhos vendados e chutar todas as resposta. Assumindo que a prova **vale 10 pontos e a nota de corte seja 5**, obtenha a probabilidade deste candidato **acertar 5 quest√µes** e tamb√©m a probabilidade deste candidato **passar para a pr√≥xima etapa do processo seletivo**.

### Qual o n√∫mero de ensaios ($n$)?

```python
n = 10
```

### Os ensaios s√£o independentes?

Sim. A op√ß√£o escolhida em uma quest√£o n√£o influencia em nada a op√ß√£o escolhida em outra quest√£o.

### Somente dois resultados s√£o poss√≠veis em cada ensaio?

Sim. O candidato tem duas possibilidades, ACERTA ou ERRAR uma quest√£o. 

### Qual a probabilidade de sucesso ($p$)?

```python
numero_alternativas_por_questao = 3
p = 1 / numero_alternativas_por_questao
p
```

### Qual a probabilidade de fracasso ($q$)?

```python
q = 1 - p
q
```

### Qual o total de eventos que se deseja obter sucesso ($k$)?

```python
k = 5
```

### Solu√ß√£o 1

```python
# probabilidade de acertar 5 quest√µes
probabilidade = (comb(n, k) * (p ** k) * (q ** (n-k)))
probabilidade * 100
```

### Importando bibliotecas
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.binom.html

```python
from scipy.stats import binom
```

### Solu√ß√£o 2

```python
probabilidade = binom.pmf(k, n, p)
probabilidade * 100
```

### Obter a probabilidade do candidato passar

### $$P(acertar \geq 5) = P(5) + P(6) + P(7) + P(8) + P(9) + P10)$$

```python
binom.pmf([5, 6, 7, 8, 9], n, p).sum() * 100
```

```python
(1 - binom.cdf(4, n, p)) * 100
```

## ‚úÖ 2.3 Distribui√ß√£o Normal
***

A distribui√ß√£o normal √© uma das mais utilizadas em estat√≠stica. √â uma distribui√ß√£o cont√≠nua, onde a distribui√ß√£o de frequ√™ncias de uma vari√°vel quantitativa apresenta a forma de sino e √© sim√©trica em rela√ß√£o a sua m√©dia.

![Normal](https://caelum-online-public.s3.amazonaws.com/1178-estatistica-parte2/01/img001.png)

### Caracter√≠sticas importantes

1. √â sim√©trica em torno da m√©dia;

2. A √°rea sob a curva corresponde √† propor√ß√£o 1 ou 100%;

3. As medidas de tend√™ncia central (m√©dia, mediana e moda) apresentam o mesmo valor;

4. Os extremos da curva tendem ao infinito em ambas as dire√ß√µes e, teoricamente, jamais tocam o eixo $x$;

5. O desvio padr√£o define o achatamento e largura da distribui√ß√£o. Curvas mais largas e mais achatadas apresentam valores maiores de desvio padr√£o;

6. A distribui√ß√£o √© definida por sua m√©dia e desvio padr√£o;

7. A probabilidade sempre ser√° igual √† √°rea sob a curva, delimitada pelos limites inferior e superior.

# $$f(x) = \frac{1}{\sqrt{2\pi\sigma}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

Onde:

$x$ = vari√°vel normal

$\sigma$ = desvio padr√£o

$\mu$ = m√©dia

A probabilidade √© obtida a partir da √°rea sob a curva, delimitada pelos limites inferior e superior especificados. Um exemplo pode ser visto na figura abaixo.

![alt text](https://caelum-online-public.s3.amazonaws.com/1178-estatistica-parte2/01/img002.png)


Para obter a √°rea acima basta calcular a integral da fun√ß√£o para os intervalos determinados. Conforme equa√ß√£o abaixo:

# $$P(L_i<x<L_s) = \int_{L_i}^{L_s}\frac{1}{\sqrt{2\pi\sigma}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

Onde:

$x$ = vari√°vel normal

$\sigma$ = desvio padr√£o

$\mu$ = m√©dia

$L_i$ = limite inferior

$L_s$ = limite superior

### Tabelas padronizadas

As tabelas padronizadas foram criadas para facilitar a obten√ß√£o dos valores das √°reas sob a curva normal e eliminar a necessidade de solucionar integrais definidas.

Para consultarmos os valores em uma tabela padronizada basta transformarmos nossa vari√°vel em uma vari√°vel padronizada $Z$.

Esta vari√°vel $Z$ representa o afastamento em desvios padr√µes de um valor da vari√°vel original em rela√ß√£o √† m√©dia.

# $$Z = \frac{x-\mu}{\sigma}$$

Onde:

$x$ = vari√°vel normal com m√©dia $\mu$ e desvio padr√£o $\sigma$

$\sigma$ = desvio padr√£o

$\mu$ = m√©dia

### Construindo tabela normal padronizada
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

tabela_normal_padronizada = pd.DataFrame(
    [], 
    index=["{0:0.2f}".format(i / 100) for i in range(0, 400, 10)],
    columns = ["{0:0.2f}".format(i / 100) for i in range(0, 10)])

for index in tabela_normal_padronizada.index:
    for column in tabela_normal_padronizada.columns:
        Z = np.round(float(index) + float(column), 2)
        tabela_normal_padronizada.loc[index, column] = "{0:0.4f}".format(norm.cdf(Z))

tabela_normal_padronizada.rename_axis('Z', axis = 'columns', inplace = True)

tabela_normal_padronizada
```

## <font color=green>3.1 Popula√ß√£o e Amostra</font>

### Popula√ß√£o
Conjunto de todos os elementos de interesse em um estudo. Diversos elementos podem compor uma popula√ß√£o, por exemplo: pessoas, idades, alturas, carros etc.

Com rela√ß√£o ao tamanho, as popula√ß√µes podem ser limitadas (popula√ß√µes finitas) ou ilimitadas (popula√ß√µes infinitas).

### Popula√ß√µes finitas

Permitem a contagem de seus elementos. Como exemplos temos o n√∫mero de funcion√°rio de uma empresa, a quantidade de alunos em uma escola etc.

### Popula√ß√µes infinitas

N√£o √© poss√≠vel contar seus elementos. Como exemplos temos a quantidade de por√ß√µes que se pode extrair da √°gua do mar para uma an√°lise, temperatura medida em cada ponto de um territ√≥rio etc.

<font color=red>Quando os elementos de uma popula√ß√£o puderem ser contados, por√©m apresentando uma quantidade muito grande, assume-se a popula√ß√£o como infinita.</font>.

### Amostra
Subconjunto representativo da popula√ß√£o.

Os atributos num√©ricos de uma popula√ß√£o como sua m√©dia, vari√¢ncia e desvio padr√£o, s√£o conhecidos como **par√¢metros**. O principal foco da infer√™ncia estat√≠stica √© justamente gerar estimativas e testar hip√≥teses sobre os par√¢metros populacionais utilizando as informa√ß√µes de amostras.

## <font color=green>3.2 Quando utilizar uma amostra?</font>
***

### Popula√ß√µes infinitas

O estudo n√£o chegaria nunca ao fim. N√£o √© poss√≠vel investigar todos os elementos da popula√ß√£o.

### Testes destrutivos

Estudos onde os elementos avaliados s√£o totalmente consumidos ou destru√≠dos. Exemplo: testes de vida √∫til, testes de seguran√ßa contra colis√µes em autom√≥veis.

### Resultados r√°pidos

Pesquisas que precisam de mais agilidade na divulga√ß√£o. Exemplo: pesquisas de opini√£o, pesquisas que envolvam problemas de sa√∫de p√∫blica.

### Custos elevados

Quando a popula√ß√£o √© finita mas muito numerosa, o custo de um censo pode tornar o processo invi√°vel.
